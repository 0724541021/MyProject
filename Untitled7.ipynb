```python
import pandas as pd
import numpy as np
df = pd.read_csv(
    r"C:\Users\Maingo Israel\OneDrive\Desktop\Masters Thesis\Top40_data_Analysis.csv",
    sep=";",                  # important for semicolon-separated values
    parse_dates=['Date']     # convert Date column to datetime
)
# Preview the first few rows
print(df.head())
```

```python
df.info()
```

```python
df.info()
df['Return'] = np.log(df['Price']).diff()*100
df[['Date','Price','Return']].head()
df
```

```python
df = df.iloc[1:]  # keeps all rows from index 1 onwards
#df = df.reset_index(drop=True)  # optional: reset index to start from 0
df
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(df['Date'], df['Return'])
plt.title("Closing price plot over time")
plt.xlabel("Year")
plt.ylabel("Closing Price")
plt.show()
```

```python
# Compute summary statistics
summary = df['Return'].describe()

# Round to 4 decimal places
summary_rounded = summary.round(5)

print(summary_rounded)
```

```python
# Skewness
skew = df['Return'].skew()

# Kurtosis
kurt = df['Return'].kurtosis()  # Fisherâ€™s definition: normal = 0

print("Skewness:", round(skew, 5))
print("Kurtosis:", round(kurt, 5))
```

```python
import seaborn as sns

# --------------------------
# 2. Create figure with 4 plots
# --------------------------
fig, axs = plt.subplots(2, 2, figsize=(14,10))  # 2x2 grid

# --- Time series plot ---
axs[0,0].plot(df['Date'], df['Return'], color='blue')
axs[0,0].set_title("Time Series of Returns", fontsize=12)
axs[0,0].set_xlabel("Date")
axs[0,0].set_ylabel("Return (%)")

# --- Histogram with density line ---
sns.histplot(df['Return'], bins=50, kde=True, color='green', ax=axs[0,1])
axs[0,1].set_title("Histogram with Density", fontsize=12)
axs[0,1].set_xlabel("Return (%)")
axs[0,1].set_ylabel("Frequency")

# --- Density plot ---
sns.kdeplot(df['Return'], fill=True, color='orange', ax=axs[1,0])
axs[1,0].set_title("Density Plot", fontsize=12)
axs[1,0].set_xlabel("Return (%)")
axs[1,0].set_ylabel("Density")

# --- Box-Whisker plot ---
sns.boxplot(x=df['Return'], color='pink', ax=axs[1,1])
axs[1,1].set_title("Box-Whisker Plot", fontsize=12)
axs[1,1].set_xlabel("Return (%)")

# --------------------------
# 3. Adjust layout
# --------------------------
plt.tight_layout()
plt.show()


```
